```python
for i in non_en_list:
  df['comment'][i] = emoji.demojize(df['comment'][i])
  filter_str = []
  start = False
  for char in df['comment'][i]:
    if char == ':' and start:
      start = False
      filter_str.append(char)
    elif char == ':':
      start = True
    if start:
      filter_str.append(char)

  df['comment'][i] = ''.join(filter_str)
  df['comment'][i] = emoji.emojize(df['comment'][i])
  print(df['comment'][i])
```
